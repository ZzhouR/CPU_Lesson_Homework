# 第一题

```python
import pandas as pd
from matplotlib import pyplot as plt

# 读取 Titanic.csv 文件
titanic_data = pd.read_csv("Titanic.csv")

# 按生还与否和性别分组计数
grouped = titanic_data.groupby(['Survived', 'Sex']).size().unstack()

# 绘制柱状图
grouped.plot(kind='bar', figsize=(10,8))
plt.xlabel('Survived', fontsize=16)
plt.ylabel('Count', fontsize=16)
plt.show()
```

# 第二题

```python
import pandas as pd

# 读取数据
df = pd.read_csv('titanic.csv')

# 查看各舱位总人数
pclass_counts = df['Pclass'].value_counts().sort_index()
print("各舱位总人数：")
print(pclass_counts)

# 查看各舱位幸存者人数
survived_counts = df.groupby('Pclass')['Survived'].sum()
print("\n各舱位幸存者人数：")
print(survived_counts)

# 计算各舱位幸存率
survival_rate = survived_counts / pclass_counts
print("\n各舱位幸存率：")
print(survival_rate)
```

```python
# 统计各舱位幸存者和遇难者年龄缺失情况
missing_age = df.groupby(['Pclass', 'Survived'])['Age'].apply(lambda x: x.isnull().sum())
print("\n各舱位幸存者/遇难者年龄缺失数：")
print(missing_age)

# 统计各舱位幸存者和遇难者总人数
total_by_group = df.groupby(['Pclass', 'Survived']).size()
print("\n各舱位幸存者/遇难者总人数：")
print(total_by_group)

# 计算各舱位幸存者资料完整率
complete_rate = 1 - (missing_age / total_by_group)
print("\n各舱位幸存者/遇难者年龄资料完整率：")
print(complete_rate)
```

通过统计可以发现：

一等舱幸存者和资料完整率都偏高，故“可以证明一等舱活下来的人更多，所以留下的资料就更全”这种说法有一定道理

# 第三题

```python
# Titanic 数据集 EDA
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# 读取数据
titanic = pd.read_csv('titanic.csv')

# 查看基本信息
display(titanic.head())
display(titanic.describe())
display(titanic.info())

# 幸存与否与性别、舱位关系
sns.countplot(x='Survived', hue='Sex', data=titanic)
plt.title('幸存与否与性别关系')
plt.show()

sns.countplot(x='Survived', hue='Pclass', data=titanic)
plt.title('幸存与否与舱位关系')
plt.show()

# 年龄分布与幸存关系
sns.histplot(data=titanic, x='Age', hue='Survived', bins=30, kde=True)
plt.title('年龄分布与幸存关系')
plt.show()
```

```python
# Diabetes 数据集 EDA
diabetes = pd.read_csv('diabetes.csv')

# 查看基本信息
display(diabetes.head())
display(diabetes.describe())
display(diabetes.info())

# Outcome（是否患糖尿病）与年龄、BMI关系
sns.boxplot(x='Outcome', y='Age', data=diabetes)
plt.title('糖尿病与年龄关系')
plt.show()

sns.boxplot(x='Outcome', y='BMI', data=diabetes)
plt.title('糖尿病与BMI关系')
plt.show()

# 各特征与Outcome的均值对比
display(diabetes.groupby('Outcome').mean())
```

```python
# Stroke 数据集 EDA
stroke = pd.read_csv('stroke.csv')

# 查看基本信息
display(stroke.head())
display(stroke.describe())
display(stroke.info())

# 是否中风与性别、年龄关系
sns.countplot(x='stroke', hue='gender', data=stroke)
plt.title('中风与性别关系')
plt.show()

sns.histplot(data=stroke, x='age', hue='stroke', bins=30, kde=True)
plt.title('年龄分布与中风关系')
plt.show()

# 各特征与中风的均值对比
display(stroke.groupby('stroke').mean(numeric_only=True))
```

## 简要发现与结论
- Titanic：女性和一等舱乘客幸存率更高，年龄较小的乘客幸存概率略高。
- Diabetes：糖尿病患者的平均年龄和BMI更高，部分特征（如血糖、胰岛素）也有明显差异。
- Stroke：中风患者年龄明显偏大，部分特征（如高血压、心脏病）均值更高。
